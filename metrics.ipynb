## Метрики в задачах классификации:
* Accuracy (верность) - доля примеров, на которых модель выдаёт правильный результат. Метрика, практически не используемая на практике в силу своей бесполезности на несбалансированных данных:

$$ Accuracy = \dfrac{TP + TN}{TP + TN + FP + FN}$$

* Для визуализации ошибок модели часто используется *confusion matrix* или *матрица ошибок*, представляющая TPR, TNR, FNR и FPR в своих ячейках, при том представление может быть также через количество примером, так и относительным:

* Precision (точность) и recall (полнота):

$$precision = \dfrac{TP}{TP + FP}$$

$$recall = \dfrac{TP}{TP + FN}$$

<img src='images/precision_recall.png'>

Пусть существует обученная модель, которая выдала результат классификации для двух или более классов. В таком случае *Precision* для каждого класса можно интерпретировать как долю правильных ответов, предсказанных моделью как этот класс. *Recall* в свою очередь показывает, сколько из всех примеров данного класса были распознаны верно. *Precision* и *Recall* применимы в условиях несбалансированных выборок.

* На основе precision и recall строится метрика *$F_\beta\,\text{score} $*, которая позволяет комбинировать *precision* и *recall*.

$$F_{\beta} = (1 + \beta^2)\dfrac{precision * recall}{\beta^2 * precision + recall} $$

Параметр $\beta$ позволяет изменять баланс между *precision* и *recall*. В случае, когда $\beta < 1$, то $F_\beta$ становится более чувствительной к $precision$, когда $\beta > 1$, то метрика чувствительна к *recall*. В предельных случаях, когда $\beta = 0$ или $\beta \rightarrow +\infty$, то $F_\beta$ вырождается в *precision* и *recall* соответственно. Так как $F_\beta$ использует комбинацию *precision* и *recall*, то она будет максимальная и равна 1, когда *precision* и *recall* равны 1 и минимальны, если *precision* и *recall* равны нулю.

* В случае задачи многоклассовой классификации одиночное значение *precision*, *recall* и других метрик может быть рассчитаны с использованием следующей из техник:
    * Macro average - для каждого из классов посчитать интересующую метрику и затем усреднить их
    * Micro average - посчитать суммарное значение TP, FP, TN, FN независимо от класса и подставить их в выражение для соответствующей из метрик.

* Ряд информативных метрик может быть получен в случаях, когда классификатор вместо целевой метки выдаёт вероятность принадлежности к какому-то классу. Различные значения порога при принятии решения о принадлежности к какому-либо классу позволяет анализировать результат обучения модели.  

Одним из наиболее распространённых таких параметров является *AUC-ROC* или *ROC AUC* — площадь (Area Under Curve) под кривой ошибок (*Receiver Operating Characteristic curve*). Кривая ошибок представляется в координатах *True Positive Rate (TPR)* (*TPR* в данном случае эквивалентна *Recall*) и *False Positive Rate (FPR)* (*Fall-Out*) в виде линии от (0,0) до (1,1):

$$TPR = recall = \dfrac{TP}{TP + FN}$$

$$FPR = \dfrac{FP}{FP + TN}$$

В случае, когда классификатор является идеальным, *ROC AUC=1*, в случае когда классификатор предсказывает случайные значения, его кривая будет стремиться к прямой от (0,0) до (1,1) и будет равен *ROC AUC=0,5*.



## Метрики в задачах регрессии

*__Среднеквадратичная ошибка__* (MSE - mean squared error):
$$\Large \text{MSE} = \dfrac{1}{l} \sum_{i=1}^{l} \left( \hat{y}_i - y_i \right)^2   $$

*__Средняя абсолютная ошибка__* (MAE - mean absolute error):
$$\Large \text{MAE} = \dfrac{1}{l} \sum_{i=1}^{l} \left| \hat{y}_i - y_i\right|   $$

*__Коэффициент $R^2$__* или *__коэффициент детерминации__*:
$$\Large R^2 =1 - \dfrac{\sum_{i=1}^{l}(\hat{y}_i - y_i)^2}{\sum_{i=1}^{l}(y_i - \bar{y}_i)^2}  $$,

где $\bar{y}_i = \dfrac{1}{l}\sum_{i=1}^{l}y_i $ - среднее значение целевого признака.
Метрика $R^2$ показывает, какую долю дисперсии данных объясняет модель. Фактически её можно интерпретировать как нормированную среднеквадратичную ошибку. Максимальное значение $R^2=1$ покзывает, что модель хорошо объясняет данные, а в случае $R^2=0$ предсказания модели близки к константным.
