# <center> Линейные модели: линейная и логистическая регрессия </center>

Линейные модели активно исследуются на протяжении многих лет. В линейных моделях предполагается, что функция регресси $ \Bbb{E}(Y|X) $ является линейной комбинацией входных векторов $ X_i $. Они являются простыми, интерпретируемыми моделями, позволяющие оценить важность и влияние каждого из признаков на целевой параметр. Для некоторых задач линейные модели позволяют превзойти простые нелинейные модели, особенно в случаях малого объема обучающей выборки, разреженности данных или данных с низким отношением сигнал/шум $SNR$. Использование линейных моделей расширяется возможностью применения преобразований входных данных перед подачей на вход модели. 

В качестве входных данных для модели могут быть:
* Числовые представления данных: рост, вес, зарплата и др.
* Трансформации числовых представлений данных: применение функций логарифмирования, квадратного корня или возведения и др.
* Расширение базиса: примером такого преобразования является полиномиальное представление входных данных $ x_2=x_1^2, x_3=x_1^3$
* Кодированное представление данных: представление данных с применением какого-нибудь вида кодирования. Примером кодирования является One-Hot encoding или Label encoding.
* Зависимость от других данных: $ x_3 = x_1 * x_2 $.

В общем случае не имеет значения какая форма зависимости будет у признаков - линейные модели являются линейными относительно параметров модели, а не входного представления данных.



## Линейная регрессия

Для заданного набора данных $ \{x_i, y_i\},\,i=1\ldots l $ будем рассматривать поиск целевого признака $ y_i $ как линейную зависимость от набора признаков $x_i$:

$$ \Large \vec{y} = \sum_{i=0}^{m}w_i x_i = \vec{w}^T \vec{x} $$

Зададим модель $ \overrightarrow{y} $ в следующем виде:

$$\large \overrightarrow{y} = \mathbf{X}\,\overrightarrow{w} + \epsilon,$$

* $\overrightarrow{y} \in \mathbb{R}^n$ - объясняемый или целевой признак
* $ w $ - вектор параметров модели или веса модели;
* $\mathbf{X}$ - матрица наблюдений и признаков размерности $ l $ строк на $ d+1 $ столбцов, включая фиктивную колонку единиц слева;
* $\varepsilon$ - случайная переменная, соответствующая случайной, непрогнозируемой ошибке модели.

Перепишем предсказания целевого признака моделью для каждого из элементов матрицы наблюдений:
$$ y_i = \sum_{j=0}^{d}w_j\,X_{ij} + \epsilon_j$$ 

Линейная регрессия должна удовлетворять условиям <a id='MarkovGauss'> т. Маркова-Гаусса </a>:

1. Моделируемый признак $ y $ линейно зависит от $ x $:

$$ \Large \vec{y} = \vec{w}^T x + \vec{\varepsilon} $$

2. Независимость наблюдений $ \{x_i, y_i\} $ распределены одинаково и независимо (i.i.d. - independent and indentically distributed).

3. Между признаками не должно существовать линейной зависимости. В случае представления признаков в виде матрицы $\mathbf{\mathit{X}}$ размерностью $ l \times (d+1) $, где $ d $ - количество признаков, ранг матрицы $ \boldsymbol{\mathit{X}} $ 

\begin{equation}
\Large \mathbf{\mathit{X}} = 
\begin{bmatrix}
    1 & x_{12} & x_{13} & \dots  & x_{1(d+1)} \\
    1 & x_{22} & x_{23} & \dots  & x_{2(d+1)} \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    1 & x_{l2} & x_{l3} & \dots  & x_{l(d+1)}
\end{bmatrix} 
\end{equation}

должен быть равен $ d+1 $:

$$ \Large \text{rang} X = d + 1 $$

Наличие линейной комбинации между признаками называется *мультиколлинеарностью*.

4. Мат. ожидание ошибок $ \vec{\varepsilon} $ не зависит от $ x $:

$$ \Large \Bbb{E}(\varepsilon | x) = 0 $$

5. [Гомоскедастичность](https://en.wikipedia.org/wiki/Homoscedasticity) случайных ошибок - дисперсия случайных ошибок одинакова и конечна:

$$ \Large \forall  i: \mathrm{Var}(\varepsilon_i)=\sigma^2 < \infty $$

Оценка $\hat{w}_i$ весов $w_i$ называется линейной, если:
$$\large \hat{w}_i = w_{1i}y_1 + w_{2i}y_2 + \ldots + w_{li}y_l,$$

где $\forall k\;w_{ki}$ зависит только от наблюдаемых данных $ \mathbf{X} $ и почти наверняка нелинейно. Так как решением задачи поиска оптимальных весов будет именно линейная оценка, то и модель называется *линейной регрессией*. 

Оценка $\hat{w}_i$ называется несмещенной тогда, когда матожидание оценки равно реальному, но неизвестному значению оцениваемого параметра:

$$ \large \Bbb{E}[\hat{w}_i] = w_i $$.

Один из способов вычислить значения параметров модели является **метод наименьших квадратов (МНК)**, который минимизирует среднеквадратичную ошибку между реальным значением зависимой переменной и прогнозом, выданным моделью:

$$ \large \mathcal{L}(X, \vec{y}, \vec{w}) = \dfrac{1}{2n}\sum_{i=1}^{n} \left( y_i - \vec{w}^T \vec{x}_i \right)^2 = \\ 
\large = \dfrac{1}{2n}\|\vec{y} - X \vec{w} \|_{2}^{2} = \\ 
\large = \dfrac{1}{2n}(\vec{y} - X \vec{w})^T(\vec{y} - X \vec{w}) $$

Для решения данной оптимизационной задачи необходимо вычислить производные по параметрам модели, приравнять их к нулю и решить полученные уравнения относительно $\overrightarrow{w}$:

$$\large \dfrac{\partial \mathcal{L}}{\partial \vec{w}} = 0 \iff \vec{w} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T \vec{y}$$.

Имея в виду все определения и условия описанные выше, мы можем утверждать, опираясь на теорему [Маркова-Гаусса](https://ru.wikipedia.org/wiki/%D0%A2%D0%B5%D0%BE%D1%80%D0%B5%D0%BC%D0%B0_%D0%93%D0%B0%D1%83%D1%81%D1%81%D0%B0_%E2%80%94_%D0%9C%D0%B0%D1%80%D0%BA%D0%BE%D0%B2%D0%B0), что оценка МНК является лучшей оценкой параметров модели, среди всех линейных и несмещенных оценок, то есть обладающей наименьшей дисперсией.

**Метод максимального правдоподобия.** 

Можно показать, что использование МНК с учётом требований т. Маркова-Гаусса позволяет максимизировать максимальное правдоподобие.

Почему необходимо минимизировать среднеквадратичную ошибку? В случае изменения минимизируемого значения происходит нарушение условий т. Маркова-Гаусса, и оценки перестают быть лучшими среди линейных и несмещенных оценок.

Случайные ошибки берутся из центрированного нормального распределения и с точки зрения вероятности можно записать:

$$\large \varepsilon_i \sim \mathcal{N}(0, \sigma^2) $$.

Вероятность предсказания моделью целевого признака $ y_i $ можно представить в следующем виде:

$$\Large  \mathcal{p}(y_i | \mathbf{X}, \overrightarrow{w}) = \sum_{j=1}^{m} w_j \mathbf{X}_{ij} + \mathcal{N}(0, \sigma^2) = \mathcal{N}\left(\sum_{j=1}^{m} w_j \mathbf{X}_{ij}, \sigma^2\right)$$.

Так как примеры [*i.i.d.*, а ошибки не скоррелированы](#MarkovGauss), то полное правдоподобие данных $ \mathcal{L} $ запишем как произведение функций плотности $p(y_i)$:

$$ \Large \mathcal{L} = p\left(\vec{y}|\mathbf{X}, \vec{w}\right) = \prod_{i=1}^{n} \mathcal{N}\left(\sum_{j=1}^{m} w_j \mathbf{X}_{ij}, \sigma^2\right) $$

При вычислении произведения вероятностей возниакет проблема оперирования с малыми числами. Так как каждое значение вероятности меньше 1, то произвдение малых величин будет стремиться к нулю. Задача максимизации правдоподобия эквивалетно можно быть записана через логарифм, что позволит перейти от произведения к сумме:

$$\Large \log{\mathcal{p}}(\overrightarrow{y} | \mathbf{X}, \overrightarrow{w}) = \log \prod_{i=1}^{n} \mathcal{N}(\sum_{j=1}^{m} w_j \mathbf{X}_{ij}, \sigma^2) = \\ 
\Large = \sum_{i=1}^{n} \log{\mathcal{N}} ( \sum_{j=1}^{m} w_j \mathbf{X}_{ij}, \sigma^2) = 
-\dfrac{n}{2}\log{2\pi\sigma^2} - \dfrac{1}{2 \sigma^2}\sum_{i=1}^{n}\left(y_i - \overrightarrow{w}^T \overrightarrow{x}_i\right)^2$$ 

Нахождение гипотезы максимального правдоподобия эквивалентна максимизации выражения $ \log{\mathcal{p}}(\vec{y} | \mathbf{X}, \vec{w})$ и это то же, что и максимизация логарифма правдоподобия. Обратите внимание, что при максимизации функции по какому-то параметру можно выкинуть все члены, не зависящие от этого параметра:

$$\large \hat{w} = arg\,max_{w} \log{p}(\overrightarrow{y} | \mathbf{X}, \overrightarrow{w}) = \\ 
\large = arg\,max_{w} -\dfrac{n}{2}\log{2\pi\sigma^2} - \dfrac{1}{2 \sigma^2}\sum_{i=1}^{n}\left(y_i - \overrightarrow{w}^T \overrightarrow{x}_i\right)^2 \\ 
\large = arg\,max_{w} - \dfrac{1}{2 \sigma^2}\sum_{i=1}^{n}\left(y_i - \overrightarrow{w}^T \overrightarrow{x}_i\right)^2 = \\ 
\large = arg\,max_{w} -\mathcal{L}(\mathbf{X},\overrightarrow{y}, \overrightarrow{w}) $$

То есть максимизация правдоподобия $\mathcal{L} $ при соблюдении условий [т.Маркова-Гаусса](#MarkovGauss) эквивалентна задаче минимизации среднеквадратичной ошибки (СКО). Таким образом, вид функции ошибки является следствием нормального распределения ошибок распределена нормально.

Эти требования накладывают большие ограничения применимости линейных моделей.

**Разложение ошибки на смещение и разброс (Bias-variance decomposition).** 

При анализе ошибки линейной регрессии воспользуемся её свойствами:
* истинное значение целевого признака $ \vec{y} $ складывается из некоторой детерминированной функции $f(\vec{x})$ и случайной ошибки $\vec{\varepsilon}$: $ y = f(\vec{x}) + \vec{\epsilon} $
* ошибка распределена нормально с центром в нуле и некоторым значением дисперсии или разбросом: $ \varepsilon \sim \mathcal{N}\left(0, \sigma^2\right)$
* истинное значение целевой переменной $ \vec{y} $ также распределено нормально: $ y \sim \mathcal{N}(f(\vec{x}), \sigma^2)$
* решаемая задача - приближение некоторой детерминированной, но неизвестной функции $f(\vec{x})$ линейной функцией от регрессоров $\hat{f}(\vec{x})$, которая является точечной оценкой функции $ f $ в пространстве функций семейства линейных функций, т.е. случайной переменной, которая имеет среднее значение и дисперсию.

На основе этого можно разложить ошибку в точке $ \vec{x}$ следующим образом:

$$\large Err(\overrightarrow{x}) = \mathbb{E}\left[ \left( y - \hat{f}(\overrightarrow{x}) \right)^2 \right] = \\
\large = \mathbb{E} \left[ y^2 \right] + \mathbb{E}\left[ \left( \hat{f}(\overrightarrow{x}\right)^2 \right] - 2\,\mathbb{E}\left[y\hat{f}(\overrightarrow{x})\right] = \\
\large = \mathbb{E} \left[ y^2 \right] + \mathbb{E} \left[ \hat{f}^2 \right] - 2\,\mathbb{E}\left[y\hat{f}\right]$$

Для наглядности опустим обозначение аргумента функций. Рассмотрим каждый член в отдельности, первые два расписываются по соотношению $ Var(z) = \mathbb{E}[z^2] - \mathbb{E}[z]^2 $:

$$\large \mathbb{E}[y^2] = Var(y) + \mathbb{E}[y]^2 = \sigma^2 + f^2 $$

$$ \large \mathbb{E}[\hat{f}^2] = Var\left(\hat{f}\right) + \mathbb{E}\left[\hat{f}\right]^2 $$

Учитывая, что ошибка и целевая переменная независимы друг от друга, запишем последнее слагаемое:

$$ \large \mathbb{E}[y\hat{f}] = \mathbb{E} \left[ (f+\epsilon) \hat{f} \right] 
\large = \mathbb{E} \left[ f \hat{f} \right] + \mathbb{E}[\epsilon\hat{f}]
\large = f \mathbb{E} \left[ \hat{f} \right] + \mathbb{E} [\epsilon] \mathbb{E} \left[ \hat{f} \right] = f\mathbb{E}\left[\hat{f}\right] $$

Объединяя результаты, получим:

$$ \large Err(\overrightarrow{x}) = \mathbb{E}\left[ \left( y - \hat{f}(\overrightarrow{x}) \right)^2 \right] = \\
= \large \sigma^2 + f^2 + Var(\hat{f}) + \mathbb{E}\left[\hat{f}\right]^2 - 2f\mathbb{E}\left[\hat{f}\right] = \\
= \large (f - \mathbb{E}\left[\hat{f}\right])^2 + Var\left( \hat{f} \right)^2 + \sigma^2$$.

Получим итоговое выражение:
$$\Huge Err(\overrightarrow{x}) = Bias\left(\hat{f}\right)^2 + Var\left(\hat{f}\right) + \sigma^2$$

Итак, мы достигли цели всех вычислений, описанных выше, последняя формула говорит нам, что ошибка прогноза любой модели вида $y = f(\overrightarrow{x}) + \epsilon $ складывается из:

* квадрата смещения $\large Bias\left(\hat{f}\right)$ – средняя ошибка по всевозможным наборам данных;
* дисперсии $\large Var\left(\hat{f}\right) $ - вариативность ошибки, то, на сколько ошибка будет отличаться, если обучать модель на разных наборах данных;
* неустранимой ошибки: $\large \sigma^2$.

Если с последней мы ничего сделать не можем, то на первые два слагаемых мы можем как-то влиять. В идеале, конечно же, хотелось бы свести на нет оба этих слагаемых (левый верхний квадрат рисунка), но на практике часто приходится балансировать между смещенными и нестабильными оценками (высокая дисперсия).

<img src="images/bias_var.png"/>

Как правило, при увеличении сложности модели (например, при увеличении количества свободных параметров) увеличивается дисперсия (разброс) оценки, но уменьшается смещение. Из-за того что тренировочный набор данных полностью запоминается вместо обобщения, небольшие изменения приводят к неожиданным результатам (переобучение). Если же модель слабая, то она не в состоянии выучить закономерность, в результате выучивается что-то другое, смещенное относительно правильного решения.

<img src="images/model_complexity.png"/>

Теорема Маркова-Гаусса как раз утверждает, что МНК-оценка параметров линейной модели является самой лучшей в классе несмещенных линейных оценок, то есть с наименьшей дисперсией. Это значит, что если существует какая-либо другая несмещенная модель тоже из класса линейных моделей, то мы можем быть уверены, что $ Var\left(\hat{f}\right) \leq Var\left(g\right) $.
