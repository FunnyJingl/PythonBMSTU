Любой созданный массив в Python обладает рядом атрибутов:

* **ndim** - размерность массива или количество осей массива

* **shape** - количество элементов для каждой из размерностей массива, представляется в виде кортежа из ndim элементов; например для двумерного массива (n, m) будет соответствовать n строкам и m столбцам

* **size** - общее количество элементов в массиве, равно произведению элементов кортежа `shape`

* **dtype** - объект типа данных элекнтов массива. В случае отсутствия объявления типа тип будет присвоен автоматически. При создании массива NumPy позволяет конвертировать встроенные типы данных в Python, а также позволяет использовать собственные типы, например `numpy.int32`, `numpy.int64` и `numpy.float64`.

* **itemsize** - размер в байтах каждого элемента массива. Например, элементы массива `numpy.float64` занимают 64/8 = 8 байт. Определение `itemsize` может быть также получено через выражение `numpy.dtype.itemsize`.

* **data** - буфер, содержащий текущие значения массива. При стандартных операциях этот аттрибут, как правило, не используется - это связано с получением элементов массива с использованием операции среза массива.

Рассмотрим различные функции, применяемые при операциях с массивами:

* [zeros(shape, dtype=float, order='C')](https://docs.scipy.org/doc/numpy/reference/generated/numpy.zeros.html) - создаёт массив с нулями определённого типа. Параметр order определяет формат хранения данных. 

* [ones(shape, dtype=None, order='C')](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ones.html) - аналогично `zeros`, но создаёт массив из единиц.

* [ndarray.fill(value)](https://docs.scipy.org/doc/numpy-1.14.1/reference/generated/numpy.ndarray.fill.html) - заполнение массива значениями `value`.

* [arange(start, stop, step, dtype=None)](https://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html) - создает массив в диапазоне от `start` до `stop` с шагом `step`.

* [linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)](https://docs.scipy.org/doc/numpy/reference/generated/numpy.linspace.html#numpy.linspace) - создаёт массив при использовании нецелого значения шага в диапазона `[start, stop)` с количеством элементов `num`. Как правило, `linspace` применяется в случае, когда шаг между значениями не является целочисленным.

**Срезы массивов**. Операция среза массива создаёт копию массива в соответствии с индексами. Индексация массива производится по правилам, которые поддерживаются встренно в Python. Общая нотация операции среза следующая: $ [start: end: step] $, где start - стартовый индекс среза, `end` - индекс последнего элемента, `step` - шаг среза. По умолчанию элемент с индексом `end` не включается в срез. Массивы также могут быть изменены с помощью операции среза.



* $ [:] $ - копия всего массива 
* $ [start: end] $ - срез массива с элементами от `start` до `end`
* $ arr[: end] $ - срез массива до элемента `end`

```python
# Slicsing arrays

a = np.arange(9)
print('Slice all elements -> ', a[:])
print('Slice first 5 elements -> ', a[:5])
print('Slice last 5 elements -> ', a[-5:])
print('Slice from 2 to 5 elements -> ', a[2: 5])
print('Slice from last 5 to last 5 elements -> ', a[-5: -2])
```

```python
# Changing slicing arrays
a = np.arange(9)
a[:5] = a[:5] + 100
print('Add 100 to first 5 elements result -> ', a[:])

a[:3] = 0
print('Set to zeros first 3 elements result -> ', a[:])
```

**Тонкости использования**. При создании и операциях с массивами стоит иметь ввиду, создаётся ли та или иная операция новый массив или оперирует со старым значением. Рассмотрим несколько примеров.

```python
a = np.arange(5)  # создаем массив [0 1 2 3 4]
b = a             # массив b теперь ссылается на данные массива a
b[1] = 100        # так как массив b ссылается на данные a, то они также меняются
print('a -> ', a)
print('b -> ', b)
```

Проверка идентичности дааных, на которые ссылаются различные объекты, реализованы в Python с помощью функции `is`.

```python
print('Являются ли данные из массива b общими для массива a?')
print(a is b)
```

```python
a = np.arange(5)  # создаем массив [0 1 2 3 4]
b = a.copy()      # массив b теперь ссылается на копию данных массива a
b[1] = 100        # так как массив b ссылается на данные a, то они также меняются
print('a -> ', a)
print('b -> ', b)
```

```python
a = np.arange(5)  # создаем массив [0 1 2 3 4]
b = a[:3]         # b  - операция среза массива a, теперь ссылается на копию данных массива a
b[1] = 100        # так как массив b ссылается на данные a, то они также меняются
print('a -> ', a)
print('b -> ', b)
```

**Изменение размерности массива.** Изменение размерности массива производится с помощью функции `numpy.reshape`. Одно из значений аргумента `shape` может принимать значение -1 - тогда изменение размерности производится по остаточному принципе и с согласованием нового размера данных. 

Массив с любой размерностью может быть преобразован в одномерный массив с помощью функции `numpy.ravel()` или `numpy.ndarray.ravel()`. 

```python
a = np.arange(16)         # create 1D array
print(a)  
print(a.shape, '\n')

b = a.reshape(4,4)        # create 2D array
print(b)
print(b.shape, '\n')

b = a.reshape(-1,1)       # create 2D array
print(b)
print(b.shape, '\n')
```

```python
b = a.reshape(1, 4, -1)   # create 3D array
print(b)
print(b.shape)

b = a.reshape(1, 4, -1).ravel()   # create 3d array and transform to 1D
print(b)
print(b.shape)

a = np.arange(5)        # создаем массив [0 1 2 3 4]
b = a.reshape(a.shape)  # b - операция среза массива a, теперь ссылается на копию данных массива a
b[1] = 100              # так как массив b ссылается на данные a, то они также меняются
print('a -> ', a)
print('b -> ', b)
```

Увеличение размерности также может быть осуществлено с использованием объекта `numpy.newaxis`:

```python
a = np.arange(25).reshape((5,5))

print(a.shape)
print(a[:, np.newaxis].shape)
print(a[:, :, np.newaxis].shape)
print(a[..., np.newaxis].shape)
print(a[np.newaxis].shape)
```

**Проверка вхождения элемента в массив**. Проверить, существует ли тот или иной элемент в массиве, можно с помощью выражения `in`:

**Объединение массивов**. Существуют методы, позволяющие объединять различные массивы при соблюдении условия равенства размерностей. Объединение может осуществляться с помощью методов `numpy.hstack` - конкатенация вдоль стоблцов, `numpy.vstack` - конкатенация вдоль строк, `numpy.concatenate` - конкатенация вдоль заданного измерения.

[SciPy](https://docs.scipy.org/doc/numpy/index.html) - Scientific Python - библиотека, которая расширяет функционал NumPy [коллекцией полезных алгоритмов](https://docs.scipy.org/doc/scipy/reference/), таких как:

В качестве примера использования библиотеки рассмотрим генерацию сигналов, состоящих из 3 синусоид и построим Фурье-спектр этого сигнала.

```python
n_samples = 50_000
x_start = 0
x_end = 50

x = np.linspace(x_start, x_end, n_samples)
y = np.sin(10 * x) + np.sin(x) + np.sin(2*x) + np.random.normal(scale=0.1, size=x.size)
```

```python
plt.figure()
plt.plot(y)
```

```python
from scipy import fftpack
```

```python
delta_t = (x_end - x_start) / n_samples  # период дискретизации

yf = fftpack.fft(y)
xf = np.linspace(0.0, 1.0/(2.0*delta_t), n_samples/2)
```

```python
# plot spectr
fig, ax = plt.subplots()
ax.plot(xf, 2.0/n_samples * np.abs(yf[:n_samples//2]))
plt.gca().set_xlim([0, 5])
plt.show()
```

# Pandas

**[Pandas](http://pandas.pydata.org)** - библиотека для анализа данных на языке Python с открытым исходным кодом.

Основными структуры данных в Pandas: классы `Series` и `DataFrame`. 

`Series` - одномерный индексированный массив данных некоторого фиксированного типа.

__Дерево решений__ – это модель, представляющая собой совокупность правил для принятия решений.

Графически её можно представить в виде древовидной структуры, где моменты принятия решений соответствуют так называемым узлам (_decision  nodes_). В узлах происходит ветвление процесса (_branching_), т.е.  деление его на так называемые ветви (_branches_) в зависимости от сделанного выбора. Конечные (или, что то же самое, терминальные) узлы называют листьями (_leafs, leafnodes_), где каждый лист – это конечный результат последовательного принятия решений.

Построение дерева основано на приросте информации в каждом узле, т.е. уменьшении энтропии Шеннона:

$$ S=-\sum_{i}^{N}{p_i\,log_2 p_i} $$

Уменьшение энтропии при разбиении по признаку _Q_ оценивается с помощью прироста информации (Information Gain - IG):

$$ IG(Q) = S_{0} - \sum_{i=1}^{q}\dfrac{N_i}{N} S_{i} $$

# Overfitting
Сгенерируем функцию синуса на отрезке $[0, 1]$, предварительно добавив к ней шум:

```python
random_seed = 90
```

```python
np.random.seed(random_seed) # set seed for reproducable data
x_sin = np.linspace(0, 1, 10)  # x values
y_sin = np.sin(x_sin * 2 * np.pi) + np.random.randn(x_sin.size) * 0.3
```

```python
np.random.seed(random_seed)
x_plot = np.linspace(0, 1, 100000)
y_test = np.sin(x_plot * 2 * np.pi) + np.random.randn(x_plot.size) * 0.3
```

```python
plt.figure(figsize=(9, 9))
plt.plot(x_plot, np.sin(x_plot * 2 * np.pi), color='r', linewidth=2.0)
plt.scatter(x_sin, y_sin, color='darkgreen')
plt.legend(['Target', 'Noisy data'])
plt.grid(linewidth=0.2)
```

Пусть будем искать представление искомой функции в полиномиальной форме:

$$ y(x, \textbf{w}) = w_0 + w_1\,x + w_2\,x + \dots + w_M\,x^M = \sum_{j=0}^{M} w_j\,x^j,$$
где $ M $ - порядок полинома, $ w_j $ коэффициенты полинома.

Функция $ y(x, \textbf{w}) $ нелинейна относительно $ x^j $, но линейна относительно коэффицентов $ w_j $. Такие модели, в которых обучаемые параметры (коэффициенты в данном случае) называются __линейными__. 

Значения коэффициентов $ w_j $ могут быть получены за счет аппроксимации полиномом исходных данных. Аппроксимая данных полиномом производится согласно уменьшению ошибки аппроксимации $ E $. Такой ошибкой часто выступает квадратичная функция потерь:

$$ E(\textbf{w}) = \dfrac{1}{2} \sum_{n=1}^{N} (y(x_n, \textbf{w}) - t_n)^2, $$ 
где $ N $ - размер данных.

Аппроксимая данных в данном случае будет производиться так, чтобы уменьшить расстояние между каждой точкой из наблюдаемых данных и значением, предсказанным моделью. Существует значение $ w^* $, которое является минимумом функции $ E(\textbf{w}) $. Тогда функция $ y(x, \textbf{w^*}) $ будет удовлетворять условию минимума среднеквадратичной ошибки. 

Когда необходимо сравнить ошибку на данных разного размера, оценивают среднеквадратичное отклонение ошибки:
$$E_{RMS} = \sqrt{\dfrac{2E\textbf(w^*)}{N}}$$

Но какое значение степени полинома $ M $ следует выбрать? Рассмотрим значения полинома $ M = 0, 1, 2, 3, 9 $.

```python
# import sklearn regressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge, LinearRegression
```

```python
m_values = np.arange(10)
y_m = {'train': {}, 
       'test': {}}
models = {}
for m in m_values:
    poly = PolynomialFeatures(degree=m)
    poly_ = poly.fit_transform(x_sin.reshape(-1, 1))
    x_plot_ = poly.fit_transform(x_plot.reshape(-1, 1))
    lg = LinearRegression()
    lg.fit(poly_, y_sin)
    models[m] = lg
    y_m['test'][m] = lg.predict(x_plot_)
    y_m['train'][m] = lg.predict(poly.fit_transform(x_sin.reshape(-1, 1)))
```

```python
plt.figure(figsize=(9, 9))
plt.plot(x_plot, np.sin(x_plot * 2 * np.pi), color='r', linewidth=2.0)
m_except = list(range(4, 9))

for m in m_values:
    if m not in m_except:
        plt.plot(x_plot, y_m['test'][m], linestyle='--')
plt.scatter(x_sin, y_sin, color='darkgreen')

plt.legend(['Target', *('M = ' + str(i) for i in m_values if i not in m_except), 'Noisy data'])
plt.grid(linewidth=0.2)
```

Вычислим $E_{RMS}$ для каждого из значений степени полинома в случае обучающей *(x_sin, y_sin)* и тестовой выборки *(x_plot, y_test)*:

```python
# define E_rms function
def e_rms(y_true, y_pred):
    return np.sqrt(np.sum((y_true - y_pred)**2)/y_true.size)

error_rms = {'train': {}, 
             'test': {}}
for m in m_values:
    error_rms['train'][m] = e_rms(y_sin, y_m['train'][m])
    error_rms['test'][m] = e_rms(y_test, y_m['test'][m])
```

И построим графики значений $ E_{RMS} $ от значения степени полинома для обучающей и тестовой выборок:

```python
plt.figure(figsize=(9, 6))
plt.scatter(m_values, np.array([v for k, v in error_rms['train'].items()]))
plt.scatter(m_values, np.array([v for k, v in error_rms['test'].items()]))
plt.legend(['Train', 'Test'])
plt.xlabel(r'Polynom degree $ m$', fontsize=20)
plt.ylabel(r'$E_{RMS}$', fontsize=20)
plt.grid(linewidth=0.3)
```

Сделаем выводы по графику:
* Ошибка на train упала до 0 в случае $ M = 9 $, модель при этом абсолютно точно аппроксимировала данные. Ошибка на тестовой выборке резко возросла.
* Качественную аппроксимацию данных обеспечивают порядки полинома от 3 до 8

Что же происходит с весами полинома при этом? Посмотрим на них:


```python
for m in m_values:
    print('Model with degree M={}'.format(m))
    for n_w, weight in enumerate(models[m].coef_):
        print('w_{} = {:.2f}'.format(n_w, weight), end=' ')
    print()
```

Значения коэффициентов возрастают с увеличением степени полинома, а модель пытается аппроксимровать шумовые значения. Это явление называется **`overfitting`**.

Обучим модель со степенью полинома 9, но увеличим обучающую выборку до 100 значений:

```python
np.random.seed(random_seed)
x_train = np.linspace(0, 1, 100)
y_train = np.sin(2 * np.pi * x_train) + np.random.randn(x_train.size) * 0.3
```

Построим график и посмотрим, как обучилась в таком случае модель:

```python
poly = PolynomialFeatures(degree=9)
poly_ = poly.fit_transform(x_train.reshape(-1, 1))
lg = LinearRegression()
lg.fit(poly_, y_train)
```

```python
plt.figure(figsize=(9, 9))
plt.plot(x_plot, np.sin(x_plot * 2 * np.pi), color='r', linewidth=2.0)
plt.scatter(x_train, y_train, color='darkgreen')

plt.plot(x_plot, lg.predict(poly.fit_transform(x_plot.reshape(-1, 1))), color='C4', linewidth=1.0, linestyle='--')

plt.legend(['Target', 'Noisy data'])
plt.grid(linewidth=0.2)
```

Увеличеие количества данных делает модель более гибкой - она приобретает способность более качественно обобщаться на новые данные (**`generalization`**). Но что делать, если возможности увеличение датасета не существует? Можно воспользоваться различными методиками, одной из которых является **регуляризация**.

Регуляризация штрафут функцию потерь во избежание увеличения значений весов:
$$ E(\textbf{w}) = \dfrac{1}{2} \sum_{n=1}^{N} (y(x_n, \textbf{w}) - t_n)^2 + \dfrac{\lambda}{2}\|\textbf{w}^2\|, $$ 
где $ \|\textbf{w}^2\| = \textbf{w}^\intercal \textbf{w} = w_0 + w_1\,x + w_2\,x + \dots + w_M\,x^M $,

$\lambda$ - коэффициент регуляризации, который показывает важность между компонентой среднеквадратичной ошибкой и величиной штрафа. 

Такой случай регуляризации в случае линейной регрессии получил своё название **Ridge Regression** (гребневая регрессия) или **метод регуляризации Тихонова**.
