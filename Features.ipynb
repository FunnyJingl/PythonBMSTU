При анализе и построении признаков выделяют следующие задачи:
* feature extraction или feature engineering – представление данных в понятный для модели вид

# Feature extraction

Задача извлечения признаков представляет из себя задачу получения числовых или иных данных из информации, собираемой в реальной жизни. Источником информации в этом случае может быть всё, что угодно: тексты, сигналы, звуки, изображения, временные ряды и др.

Для моделей наиболее понятными являются различные числовые представления признаков. Поэтому существуют различные подходы по представлению существующей информации в виде векторов признаков.

**Тексты**.
**Сигналы**. 

**Временные ряды**. 
**Геолокация**.
**Дата и время**.

**Изображения**. В случае изображения до бума нейросетей приходилось пользоваться рукотворными признаками - гистограмные моменты, текстура и характерные особенности для решаемой задачи, и различными дескрипторами - [SIFT](https://www.cs.ubc.ca/~lowe/papers/ijcv04.pdf), [SURF](https://www.vision.ee.ethz.ch/~surf/eccv06.pdf), [ORB](http://www.willowgarage.com/sites/default/files/orb_final.pdf) и другие. На текущий день для надёжного выделения более сложных признаков можно воспользоваться предобученной свёрточной нейронной сетью, на выходе которой будет вектор признаков, характеризующих изображение.
