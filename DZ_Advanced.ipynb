**Домашнее задание**

Задание будет основано на датасете  https://www.kaggle.com/harlfoxem/housesalesprediction - прогнозированию цены дома в зависимости от параметров.

Вам необходимо сделать следующие задачи:

0) Скачать данные и считать их в датафрейм.

1) Проанализировать и заполнить пропуски (N/A), если таковые имеются

2) Проанализировать каждый из признаков - удалить выбросы, преобразовать к одинаковому интервалу. Преобразуйте категориальные признаки в числовые.

3) Применить t-SNE, что посмотреть на кластеризацию данных. Будьте внимательны к размеру данных, оно может быть времязатратным.

4) Поделить выборку с в соотношении 70-30 на обучающую X_train и отложенную X_holdout. На обучающей выборке примените Stritified K-Fold и обучите в каждом случае [дерево решений](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html). 

5) Для поиска оптимальных значений гиперпараметров примените [GridSearch](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html) и обучите модель с разными вариантами глубины, например 3, 5, 9.

6) На каждой итерации Stratified K-Fold делайте предсказание по оставшейся выборке и запомните его.

7) После обучение каждой из моделей сделайте предсказание на отложенной выборке и запомните его.

8) Посчитайте во всех случаях метрику MSE на оставшейся части выборки. Сделайте выводы о разнице между моделями.

9) Если вы смогли все сделать до этого пункта, то обучите [RandomForest](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html) на всей обучающей выборке 1 раз со стандартными параметрами и сделайте предсказание на отложенной выборке. Сравните результат с деревом решений.
